## Question 2 (10 Points)
Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction.

```python
# You code here (Please add comments in the code):

# Sample text data
reviews = [
    "The movie was outstanding with brilliant performances.",
    "I found the movie boring and too long.",
    "The cinematography was breathtaking but the storyline was not good.",
    "An utterly impressive masterpiece that I highly recommend.",
    "The plot was predictable and lacked originality."
]

# Ensure you have the necessary libraries installed:
# pip install nltk sklearn gensim vaderSentiment

# Import necessary libraries
import nltk
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from gensim.models import Word2Vec
from nltk.util import ngrams
from nltk.corpus import stopwords
import pandas as pd
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')

# Tokenization and POS Tagging
tokens_pos = [(nltk.word_tokenize(review), nltk.pos_tag(nltk.word_tokenize(review))) for review in reviews]

# Bag of Words
vectorizer = CountVectorizer(stop_words='english')
bow = vectorizer.fit_transform(reviews)

bow_df = pd.DataFrame(bow.toarray(), columns=vectorizer.get_feature_names_out())
print("Bag of Words (BoW):\n", bow_df)

# TF-IDF
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf = tfidf_vectorizer.fit_transform(reviews)

tfidf_df = pd.DataFrame(tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
print("TF-IDF Features:\n", tfidf_df)

# Sentiment Lexicon (VADER)
analyzer = SentimentIntensityAnalyzer()
sentiments = [analyzer.polarity_scores(review) for review in reviews]

# N-grams (bigrams example)
bigrams = [list(ngrams(nltk.word_tokenize(review), 2)) for review in reviews]

# Word Embeddings (Word2Vec)
# Train a simple Word2Vec model with our small dataset
model = Word2Vec([nltk.word_tokenize(review.lower()) for review in reviews], vector_size=100, window=5, min_count=1, workers=2)
word_vectors = model.wv

# Example of using the word embeddings
sample_vector = word_vectors['movie']

# Displaying the features
print("Bag of Words (BoW):", bow.toarray())
print("\nSentiments (VADER):", sentiments)
print("\nBigrams:", bigrams)
print("\nSample Word Vector ('movie'):", sample_vector)

```

## Question 3 (10 points):
Use any of the feature selection methods mentioned in this paper "Deng, X., Li, Y., Weng, J., & Zhang, J. (2019). Feature selection for text classification: A review. Multimedia Tools & Applications, 78(3)."

Select the most important features you extracted above, rank the features based on their importance in the descending order.

```python
# You code here (Please add comments in the code):



from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Example dataset
reviews = ["The movie was outstanding with brilliant performances.",
           "I found the movie boring and too long.",
           "The cinematography was breathtaking but the storyline was not good.",
           "An utterly impressive masterpiece that I highly recommend.",
           "The plot was predictable and lacked originality."]

# Example target variable
sentiments = [1, 0, 0, 1, 0]  # 1 for positive, 0 for negative

# Convert text data into TF-IDF features
vectorizer = TfidfVectorizer(max_features=1000, stop_words='english')
X = vectorizer.fit_transform(reviews)
y = sentiments

# Compute mutual information between each feature and the target
mi_scores = mutual_info_classif(X, y, discrete_features='auto')

# Create a Series to view the scores and feature names
mi_scores_series = pd.Series(mi_scores, index=vectorizer.get_feature_names_out())

# Sort the features by their mutual information scores in descending order
mi_scores_sorted = mi_scores_series.sort_values(ascending=False)

print(mi_scores_sorted)
```

## Question 4 (10 points):
Write python code to rank the text based on text similarity. Based on the text data you used for question 2, design a query to match the most relevant docments. Please use the BERT model to represent both your query and the text data, then calculate the cosine similarity between the query and each text in your data. Rank the similary with descending order.

```python
!pip install sentence_transformers
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Load the BERT model
model = SentenceTransformer('bert-base-nli-mean-tokens')

# Sample text data (movie reviews)
reviews = [
    "The movie was outstanding with brilliant performances.",
    "I found the movie boring and too long.",
    "The cinematography was breathtaking but the storyline was not good.",
    "An utterly impressive masterpiece that I highly recommend.",
    "The plot was predictable and lacked originality."
]

# Updated query relevant to movie reviews
query = "I'm looking for a movie with great performances"

# Compute the embeddings for the query and reviews
query_embedding = model.encode([query])[0]
review_embeddings = model.encode(reviews)

# Compute the cosine similarities between the query and reviews
similarities = cosine_similarity([query_embedding], review_embeddings)[0]

# Rank the reviews based on their similarity to the query
ranked_indices = similarities.argsort()[::-1]  # Sort indices by similarity in descending order

# Using tabulate to format the output table, ensuring it's installed
try:
    from tabulate import tabulate
except ImportError:
    !pip install tabulate
    from tabulate import tabulate

# Prepare the ranked list for display
ranked_texts = [(reviews[i], similarities[i]) for i in ranked_indices]

# Create a table to display the rank, review text, and similarity
table = [[i+1, text, round(similarity, 4)] for i, (text, similarity) in enumerate(ranked_texts)]

print(tabulate(table, headers=['Rank', 'Review', 'Similarity'], tablefmt='orgtbl'))


```
